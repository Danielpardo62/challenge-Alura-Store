### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

##Categorias

```python
categorias= [tienda_categorias, tienda2_categorias, tienda3_categorias, tienda4_categorias]
cantidades = [tienda_cantidades, tienda2_cantidades, tienda3_cantidades, tienda4_cantidades]
print(categorias)
print(cantidades)

df_categorias= pd.DataFrame({'Tienda' : tiendas, 'categoria' : categorias, 'cantidad' : cantidades})
df_categorias
```

```python
# Aplanando df_categorias para la tabla dinámica
data_for_df = []
for i, nombre_tienda in enumerate(df_categorias['Tienda']):
    for j, categoria in enumerate(df_categorias['categoria'].iloc[i]):
        data_for_df.append({
            'Tienda': nombre_tienda,
            'categoria': categoria,
            'cantidad': df_categorias['cantidad'].iloc[i][j]
        })
df_aplanada = pd.DataFrame(data_for_df)

pt = (
    df_aplanada.pivot_table(
        index="Tienda", columns="categoria", values="cantidad", aggfunc= "sum"
    ).fillna(0)
)
pt
```

```python
pt_pct = pt.div(pt.sum(axis=1), axis=0)
pt_pct
```

```python
pt = (pt.T / pt.sum(axis=1)).T * 100
pt
```

```python
x = np.arange(len(pt.index))
bottom = np.zeros(len(pt))
for c in pt.columns:
    plt.bar(x, pt[c].values, bottom=bottom, label=c)
    bottom += pt[c].values
plt.xticks(x, pt.index); plt.ylabel("Porcentaje"); plt.title("Distribución por categoría")
plt.legend(ncols=3, fontsize=8); plt.margins(x=0.1); plt.show()
```

##Puntaje por tienda

```python
df= pd.DataFrame({'Tienda': tiendas, 'ingresos': ingresos_totales , 'valoracion': opiniones, 'envio': envio})
df['ingresos'] = df['ingresos'].round().astype('int64')

ing_min, ing_max = tabla_ingresos_totales["Ingreso Total"].min(), tabla_ingresos_totales["Ingreso Total"].max()
val_min, val_max = df_opiniones["Valoración"].min(), df_opiniones["Valoración"].max()
env_min, env_max = envio_tiendas["envio"].min(), envio_tiendas["envio"].max()

df["ing_norm"] = 0.5 if ing_max == ing_min else (df["ingresos"] - ing_min) / (ing_max - ing_min)
df["val_norm"] = 0.5 if val_max == val_min else (df["valoracion"]   - val_min) / (val_max - val_min)
df["env_norm"] = 0.5 if env_max == env_min else (env_max - df["envio"]) / (env_max - env_min)

# Pesos (ajusta según prioridad; deben sumar 1)
p_ing, p_val, p_env = 0.4, 0.4, 0.2

df["puntaje"] = p_ing*df["ing_norm"] + p_val*df["val_norm"] + p_env*df["env_norm"]

x = df["Tienda"]
y = df["puntaje"]
plt.bar(x, y, width=0.6)
plt.title("Puntaje por tienda")
plt.xlabel("Tienda")
plt.ylabel("Puntaje")
plt.margins(x=0.1)
```

## Análisis y justificación
- **Ingresos:** la tienda con mayor volumen tiende a ofrecer mayor potencial de ventas.  
- **Satisfacción del cliente:** una **calificación promedio** más alta reduce el riesgo de devoluciones e impulsa repetición de compra.  
- **Coste de envío:** un **envío promedio más bajo** mejora conversión y margen.  
- **Mix de categorías:** A pesar de existir diversas categorías en las tiendas, se analiza que las 4 tiendas tienen un comportamiento similar en las ventas por categoría. Por ello no se tiene en cuenta como caracteristica principal de las tiendas.  
- **Portafolio de productos:** la existencia de un **top producto claro** y un **bottom** poco relevante ayuda a optimizar inventario.

Se construyó un **score compuesto** normalizando métricas (ingresos ↑, rating ↑, coste envío ↓) con pesos **40% ingresos, 40% rating y 20% envío**. Esta ponderación prioriza demanda y satisfacción, sin ignorar costos logísticos.
